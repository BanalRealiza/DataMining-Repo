```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dengue_df=pd.read_csv('dataset\\dengue.csv')
```

```python
dengue_df
```

```python
dengue_df.isnull().sum()
```

```python
dengue_df.info()
```

```python
dengue_df['Month']=dengue_df['Month'].astype('category')
dengue_df['Region']=dengue_df['Region'].astype('category')
```

```python
dengue_df.info()
```

```python
dengue_df
```

<h1>INSIGHT 1</h1>

```python
top5_region_death= dengue_df.groupby ('Region') ['Dengue_Cases'].sum().sort_values(ascending=False).head(5)
top5_region_death.plot(kind='bar')
plt.title('Top 5 Region that has highest number of Dengue Cases')
plt.show()
```

<h2>The Top 5 Region that has highest number of dengue cases are Region IV-A, Region III, Region VI, NCR, and Region VII. </h2>

<h1>INSIGHT 2</h1>

```python
Top5_region_death=dengue_df.groupby('Region')['Dengue_Deaths'].sum().sort_values(ascending=False).head(5)
Top5_region_death.plot(kind='bar')
plt.title('Top 5 Region that has highest number of Death Cases')
plt.show()
```

<h2>The Top 5 Region that has highest number of death cases are NCR, Region XII, Region VI, CAR, and Region VII. </h2>

<h1>INSIGHT 3</h1>

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x=dengue_df['Dengue_Cases'], y=dengue_df['Dengue_Deaths'])
plt.xlabel("Dengue Cases")
plt.ylabel("Dengue Deaths")
plt.title("Dengue Cases vs Dengue Deaths")
plt.show()
```

<h1>INSIGHT 4</h1>

```python
yearly_cases = dengue_df.groupby("Year")["Dengue_Cases"].sum()
plt.bar(yearly_cases.index, yearly_cases.values)
plt.xlabel("Year")
plt.ylabel("Total dengue cases")
plt.title("Total dengue Cases per year")
plt.xticks(rotation=45)
plt.show()
```

<h2> Year 2019 has the highest total of dengue cases</h2>

<h1>INSIGHT 5</h1>

```python
yearly_deaths = dengue_df.groupby("Year")["Dengue_Deaths"].sum()
plt.bar(yearly_deaths.index, yearly_deaths.values)
plt.xlabel("Year")
plt.ylabel("Total Dengue Deaths")
plt.title("Total Dengue Deaths per Year")
plt.xticks(rotation=45)
plt.show()
```

<h2> Year 2016 has the highest total of dengue death</h2>

<h1>INSIGHT 6</h1>

```python
dengue_2016 = dengue_df[dengue_df['Year'] == 2016]
month_death_cases = dengue_2016 .groupby('Month')['Dengue_Deaths'].sum()
month_death_cases.plot(kind='bar')
plt.title("Dengue Death Cases per Month in year 2016")
plt.show()
```
